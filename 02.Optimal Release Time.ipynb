```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df["date_added"]=pd.to_datetime(df["date_added"],errors='coerce')
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['date_added'] = df['date_added'].fillna(method='ffill')
```

```python
from datetime import datetime

df["release_month"]=df["date_added"].dt.month
df["release_quarter"]=df["date_added"].dt.quarter
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_df=encoder.fit_transform(df[["type"]])
```

```python
encoded_df=pd.DataFrame(encoded_df,columns=encoder.get_feature_names_out(["type"]))
```

```python
df
```

```python
df=pd.concat([df.drop(columns=["type"]),encoded_df],axis=1)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

df['genres'] = df['listed_in'].str.split(', ')
mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(df['genres']), columns=mlb.classes_)
df = pd.concat([df, genre_encoded], axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
averageRating_median=df["averageRating"].median()
numVotes_median=df["numVotes"].median()
```

```python
df["result"]=((df["averageRating"] >=averageRating_median) | (df["numVotes"] >=numVotes_median)).astype(int)
```

```python
df.info()
```

```python
X=df.drop(columns=["title","director","cast","country","date_added","release_year","duration","listed_in","averageRating","numVotes","release_quarter","type_TV Show","genres","result"])
y=df["result"]
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42)
X_train_smote,y_train_smote=smote.fit_resample(X_train,y_train)
```

```python
X_train.info()
```

```python
random_forest=RandomForestClassifier()
random_forest.fit(X_train_smote,y_train_smote)
```

```python
y_pred=random_forest.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

H0=The faeture has no effect on success

H1=The feature does affect success

Chi-square test

```python
from scipy.stats import chi2_contingency
contingency_table=pd.crosstab(df['release_month'],df['result'])

chi2,p,dof,expected=chi2_contingency(contingency_table)

print(f"Chi2 Statistic:{chi2},p-value:{p}")
if p <0.05:
    print("Reject H0:release month affects success")
else:
    print("Fail to reject H0: release has no effect on success")
```

```python
# Calculate success rate for each release month
month_success = df.groupby('release_month')['result'].mean().sort_values(ascending=False)
print(month_success)

# Optional: Visualize
plt.figure(figsize=(10,5))
sns.barplot(x=month_success.index, y=month_success.values)
plt.xlabel('Release Month')
plt.ylabel('Success Rate')
plt.title('Success Rate by Release Month')
plt.show()
```

```python
# Create a pivot table for heatmap
pivot = df.pivot_table(index='release_month', columns='release_quarter', values='result', aggfunc='mean')

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Success Rate by Release Month and Quarter')
plt.xlabel('Release Quarter')
plt.ylabel('Release Month')
plt.show()
```
