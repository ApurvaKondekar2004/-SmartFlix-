```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cleaned_data.csv")
df
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df["date_added"]=pd.to_datetime(df["date_added"],errors='coerce')
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['date_added'] = df['date_added'].fillna(method='ffill')
```

```python
from datetime import datetime

df["release_month"]=df["date_added"].dt.month
df["release_quarter"]=df["date_added"].dt.quarter
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_df=encoder.fit_transform(df[["type"]])
```

```python
encoded_df=pd.DataFrame(encoded_df,columns=encoder.get_feature_names_out(["type"]))
```

```python
df
```

```python
df=pd.concat([df.drop(columns=["type"]),encoded_df],axis=1)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

df['genres'] = df['listed_in'].str.split(', ')
mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(df['genres']), columns=mlb.classes_)
df = pd.concat([df, genre_encoded], axis=1)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
averageRating_median=df["averageRating"].median()
numVotes_median=df["numVotes"].median()
```

```python
df["result"]=((df["averageRating"] >=averageRating_median) | (df["numVotes"] >=numVotes_median)).astype(int)
```

```python
df.info()
```

```python
X=df.drop(columns=["title","director","cast","country","date_added","release_year","duration","listed_in","averageRating","numVotes","release_quarter","type_TV Show","genres","result"])
y=df["result"]
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(df['result'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42)
X_train_smote,y_train_smote=smote.fit_resample(X_train,y_train)
```

```python
X_train.info()
```

```python
random_forest=RandomForestClassifier()
random_forest.fit(X_train_smote,y_train_smote)
```

```python
y_pred=random_forest.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

```python
model=XGBClassifier()
model.fit(X_train_smote,y_train_smote)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
